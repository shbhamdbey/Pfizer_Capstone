```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
my_url ='https://www.gdacs.org/xml/rss_7d.xml'
```

```python
uClient = uReq(my_url)
```

```python
page_xml = uClient.read()
```

```python
page_soup = soup(page_xml,'xml')
```

```python
print(page_xml)
```

```python

page_soup.item
```

```python
grabbing = page_soup.findAll('item')
```

```python
len(grabbing)
```

```python
grabbing[0]
```

```python
grabbing[0].find('title').text
population_list = []
```

```python
for gra in grabbing:
    for ob in gra.find_all('gdacs:population'):
        population = ob["value"]
        pop = {'Population': population}
        population_list.append(pop)
     
```

```python
print(population_list)
new_list = []
```

```python
for grab in grabbing:
    title = grab.find('title').text
    latti = grab.find('geo:lat').text
    longi = grab.find('geo:long').text
    FromDate = grab.find('gdacs:fromdate').text
    Duration = grab.find('gdacs:durationinweek').text
    EventType = grab.find('gdacs:eventtype').text
    AlertLevel = grab.find('gdacs:eventtype').text
    AlertScore = grab.find('gdacs:alertscore').text
    EPalertLevel = grab.find('gdacs:episodealertlevel').text
    EPalertScore = grab.find('gdacs:episodealertscore').text
    article = {'Title': title,
               'FromDate' : FromDate,
               'Latitude': latti,
               'Longitude': longi,
               'Duration': Duration,
               'EventType': EventType,
               'AlertLevel' : AlertLevel,
               'AlertScore' : AlertScore,
               'EPalertLevel' : EPalertLevel,
               'EPalertScore' : EPalertScore}
    new_list.append(article)

    
```

```python
print(new_list)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(new_list)
dpop = pd.DataFrame(population_list)
```
