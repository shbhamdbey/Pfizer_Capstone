### Find minimum distance (km) from tornado line to location point and then store values in location df

# Working Solution
### Formula: https://stackoverflow.com/questions/32771458/distance-from-lat-lng-point-to-minor-arc-segment 

```python
def bear(latA, lonA, latB, lonB):
    b = math.atan2( math.sin(math.radians(lonB-lonA))*math.cos(math.radians(latB)), math.cos(math.radians(latA))*math.sin(math.radians(latB)) - math.sin(math.radians(latA))*math.cos(math.radians(latB))*math.cos(math.radians(lonB-lonA)) )
    return b

def dis(latA, lonA, latB, lonB):
    d = math.acos( math.sin(math.radians(latA))*math.sin(math.radians(latB)) + math.cos(math.radians(latA))*math.cos(math.radians(latB))*math.cos(math.radians(lonB-lonA)) ) * R
    return d
```

### Calculates the distance from each tornado path to each location point. 

```python
idx = 0

for i, loc in locDF.iterrows():
    
    for j, tor in torDF.iterrows():

        R=6371; # Earth's radius in kilometers
        pi = 3.14159265359
        
        # Prerequisites for the formulas
        bear12 = bear(tor['slat'],tor['slon'],tor['p_elat'],tor['p_elon'])
        bear13 = bear(tor['slat'],tor['slon'],loc['Latitude'],loc['Longitude'])
        dis13 = dis(tor['slat'],tor['slon'],loc['Latitude'],loc['Longitude']);
        dis23 = dis(tor['p_elat'],tor['p_elon'],loc['Latitude'],loc['Longitude'])
        disAdd = dis13 + dis23
        
        diff = abs(bear13-bear12)

        if diff > pi:
            
            diff = 2 * pi - diff
        
        if diff>(pi/2): # Is relative bearing obtuse?
            
            dxa=dis13
            
        else:
            
            # Find the cross-track distance.
            dxt = math.asin( math.sin(dis13/R)* math.sin(bear13 - bear12) ) * R

            # Is p4 beyond the arc?
            dis12 = dis(torDF.at[j, 'slat'],torDF.at[j, 'slon'],torDF.at[j, 'p_elat'],torDF.at[j, 'p_elon'])
            dis14 = math.acos( math.cos(dis13/R) / math.cos(dxt/R) ) * R
            
            if dis14>dis12:
                
                dxa=dis(tor['p_elat'],tor['p_elon'],loc['Latitude'],loc['Longitude'])
                
            else:
                
                dxa=abs(dxt)
            
            df2 = pd.DataFrame({'State': locDF.at[i, 'State'], 'Capital City':locDF.at[i, 'Capital City'], 
                                'Latitude':locDF.at[i, 'Latitude'], 'Longitude':locDF.at[i, 'Longitude'], 
                                'minDist':dxa, 'Index':tor['index']}, index=[idx])
                
            locDF = locDF.append(df2, ignore_index=True)
            
            idx+=1
```
